```python
import pandas as np

dados = np.read_csv("https://raw.githubusercontent.com/talesjoabe/especiesAmeacadas/main/dados_tratados.csv")
dados.head()
```

```python
dados.drop('Unnamed: 0',axis=1, inplace=True)
dados.head()
```

# Biomas

```python
dados2 = dados.loc[:] 
dados2 = dados2.assign(Bioma=dados2.Bioma.str.split(",")).explode('Bioma')
```

```python
dados2.groupby('Bioma').size()
```

```python
indexNames = dados2[ dados2['Bioma'] == '' ].index
indexNames
```

```python
dados2.drop(indexNames , inplace=True)
dados2.dropna(axis=0, how='any', inplace=True)
```

```python
biomas = dados2.groupby('Bioma').size().reset_index(name='quantidade')
biomas.to_csv('biomas_especies_quantidade.csv')
biomas
```

# Qauntidade de especies ameacadas por Estados

```python
estados_dataset = dados.loc[:]
estados_dataset.rename(columns={"Estados de Ocorrência": "Estado_de_Ocorrencia","Fauna/Flora":"Fauna_Flora"},inplace = True)
estados_dataset = estados_dataset.assign(Estado_de_Ocorrencia=estados_dataset.Estado_de_Ocorrencia.str.split(",")).explode('Estado_de_Ocorrencia')
estados_dataset.reset_index(drop=True)
estados_dataset.head()
```

```python
indexNames = estados_dataset[ estados_dataset['Estado_de_Ocorrencia'] == '' ].index
indexNames
```

```python
estados_dataset.drop(indexNames , inplace=True)
```

```python
estados_dataset.dropna(axis=0, how='any', inplace=True)
```

```python
estados_dataset['Estado_de_Ocorrencia'].unique()
```

```python
estados_dataset['Grupo'].unique()
```

```python
especies_ameacadas=estados_dataset.groupby('Estado_de_Ocorrencia').size().reset_index(name='quantidade')
especies_ameacadas
# especies_ameacadas.to_csv('quantidade_de_especies_ameacadas_por_estado.csv')

```

# Separacao do aspecto da especie se e planta ou animal por estados em quantidade

```python
estados_dataset['Fauna_Flora'].unique()
```

```python
f = estados_dataset[estados_dataset['Fauna_Flora']=='Flora']
flora=f.groupby('Estado_de_Ocorrencia')['Fauna_Flora'].size().reset_index(name='Flora_qtd')
flora

```

```python
fa = estados_dataset[estados_dataset['Fauna_Flora']=='Fauna']
fauna=fa.groupby('Estado_de_Ocorrencia')['Fauna_Flora'].size().reset_index(name='Fauna_qtd')
fauna
```

```python
especies_ameacadas_fauna_flora_total = np.concat([especies_ameacadas,fauna["Fauna_qtd"],flora["Flora_qtd"]],axis=1)
especies_ameacadas_fauna_flora_total.to_csv('quantidade_de_especies_ameacadas_fauna_flora_total_por_estado.csv')
```
